```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Criando um conjunto de dados fictício
dados = {
    'area': [1500, 1600, 1700, 1800, 1900, 2000],
    'preco': [300000, 320000, 340000, 360000, 380000, 400000]
}

# Convertendo os dados para um DataFrame
df = pd.DataFrame(dados)

# Separando as variáveis independentes (X) e a variável dependente (y)
X = df[['area']]
y = df['preco']

# Criando o modelo de regressão linear
modelo = LinearRegression()

# Treinando o modelo com todos os dados
modelo.fit(X, y)

# Fazendo previsões para os dados
y_pred = modelo.predict(X)

# Plotando os dados e a linha de regressão
plt.scatter(X, y, color='blue', label='Dados Reais')
plt.plot(X, y_pred, color='red', label='Linha de Regressão')
plt.xlabel('Área (m²)')
plt.ylabel('Preço (R$)')
plt.title('Regressão Linear Simples')
plt.legend()
plt.show()
```
