```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os dados
url = '/content/titanic.csv'
df = pd.read_csv(url)
df = df[['Survived', 'Age']]
df.dropna(subset=['Age'], inplace=True)
```

```python
df.head()
```

```python
# Preparar os dados
X = df[['Age']]
y = df['Survived']
```

```python
# Criar e treinar o modelo
modelo = LogisticRegression()
modelo.fit(X, y)
```

```python
# Fazer uma previsão
novo_passageiro = pd.DataFrame({'Age': [90]})
chance_sobrevivencia = modelo.predict_proba(novo_passageiro)[:, 1]
print(f"Chance de sobrevivência: {chance_sobrevivencia[0] * 100:.2f}%")
```

```python
# Plotar o gráfico
idades = pd.DataFrame({'Age': range(0, 81)})
chances = modelo.predict_proba(idades)[:, 1]
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x=idades['Age'], y=chances, color='blue')
plt.xlabel('Idade (Age)')
plt.ylabel('Chance de Sobrevivência')
plt.title('Chance de Sobrevivência por Idade')
plt.grid(True)
plt.show()
```
